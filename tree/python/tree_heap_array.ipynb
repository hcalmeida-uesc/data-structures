```python
class Pessoa:
    def __init__(self, strName, intAge):
        self.name = strName
        self.age = intAge

    def __str__(self):
        return "Nome: "+self.name+" Idade: "+str(self.age)
    
    def smaller(self, value):
        return True if value == None else (self.age < value.age)
    
    def bigger(self, value):
        return True if value == None else (self.age > value.age)
```

```python
class TreeHeap:
    length:int
    def __init__(self, type="MIN"):
        self.__storage = [0]
        self.__type = type
    
    @property
    def length(self):
        return self.__storage[0]
    
    @length.setter
    def length(self,value):
        self.__storage[0] = value
        
    def isEmpty(self):
        return self.length == 0
    
    @staticmethod
    def iFather(a:int) -> int:
        return int((a-(a%2))/2)
    
    def insert(self, value):
        print(f"Inserindo {value.name}:{value.age}")
        self.__storage.append(value)
        self.length += 1
        print(f"Último {self.__storage[-1].name}:{self.__storage[-1].age}")
        self.upHeap()
        print(f"Após upHeap, último {self.__storage[-1].name}:{self.__storage[-1].age}")
        
    def remove(self):
        pass    
    
    def downHeap(self):
        pass
              
    def upHeap(self):
        if self.length > 1 :
            i = self.length
            f = self.iFather(i)
            if self.__type == "MIN":
                while(i > 1 and self.__storage[i].smaller(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    i = f
                    f = self.iFather(i)
            elif self.__type == "MAX":
                while(i > 1 and self.__storage[i].bigger(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    i = f
                    f = self.iFather(i)
            else:
                raise ValueError('Wrong Heap type. It can be MIN or MAX')
            
    # def __recursiveStr(self,node,str_final):
    #     str_final += f"{node.num} {str(node.info)}\n"
    #     if node.left != None :
    #         str_final = self.__recursiveStr(node.left,str_final)
    #     if node.right != None :
    #         str_final = self.__recursiveStr(node.right,str_final)
        
    #     return str_final
    
    # def __recursiveList(self,node,list_final):
    #     list_final.append([node.num,node.info.name,node.info.age])
    #     if node.left != None :
    #         list_final = self.__recursiveList(node.left,list_final)
    #     if node.right != None :
    #         list_final = self.__recursiveList(node.right,list_final)
        
    #     return list_final
    
    def toList(self):
        if not self.isEmpty() :
            return self.__storage[1:]
        else:
            return ["Empty Heap"]
    
    def printGraph(self):
        heap_list = self.toList()
        values = []
        for x in heap_list:
            values.append(f"{x.name}:{x.age}")
        root = build(values)
        print(root)
        
    def __str__(self) -> str:
        if not self.isEmpty() :
            heap_list = self.toList()
            str_list = ""
            for x in heap_list:
                str_list += f"{x.name}:{x.age}\n"
            return str_list
        else:
            return "Empty Heap"
```
