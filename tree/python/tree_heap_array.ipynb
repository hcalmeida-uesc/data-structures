```python
class Pessoa:
    def __init__(self, strName, intAge):
        self.name = strName
        self.age = intAge

    def __str__(self):
        return "Nome: "+self.name+" Idade: "+str(self.age)
    
    def smaller(self, value):
        return True if value == None else (self.age < value.age)
    
    def bigger(self, value):
        return True if value == None else (self.age > value.age)
```

```python
class TreeHeap:
    length:int
    def __init__(self, type="MIN"):
        self.__storage = [0]
        self.__type = type
    
    @property
    def length(self):
        return self.__storage[0]
    
    @length.setter
    def length(self,value):
        self.__storage[0] = value
        
    def isEmpty(self):
        return self.length == 0
    
    @staticmethod
    def iLeft(a:int) -> int:
        return int(a*2)
    
    @staticmethod
    def iRight(a:int) -> int:
        return int((a*2)+1)
    
    def __brother(self, a):
        return a+1
    
    @staticmethod
    def iFather(a:int) -> int:
        return int((a-(a%2))/2)
    
    def __smallerIndex(self,left,right):
        return left if right > self.length-1 or self.__storage[left].smaller(self.__storage[right]) else right
    
    def insert(self, value):
        print(f"Inserindo {value.name}:{value.age}")
        self.__storage.append(value)
        self.length += 1
        print(f"Último {self.__storage[-1].name}:{self.__storage[-1].age}")
        self.upHeap()
        print(f"Após upHeap, último {self.__storage[-1].name}:{self.__storage[-1].age}")
        
    def remove(self):
        root_info = self.__storage[1]
        self.__storage[1] = self.__storage.pop(-1)
        self.length -= 1
        self.downHeap()
        return root_info    
    
    def downHeap(self):
        if self.length > 1 :
            f = 1
            i = self.__smallerIndex(self.iLeft(f),self.iRight(f))
            if self.__type == "MIN":
                while(i < self.length-1 and self.__storage[i].smaller(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    print(self.__storage[i])
                    f = i
                    i = self.__smallerIndex(self.iLeft(f),self.iRight(f))
            elif self.__type == "MAX":
                while(i > 1 and self.__storage[i].bigger(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    i = f
                    f = self.iFather(i)
            else:
                raise ValueError('Wrong Heap type. It can be MIN or MAX')
              
    def upHeap(self):
        if self.length > 1 :
            i = self.length
            f = self.iFather(i)
            if self.__type == "MIN":
                while(i > 1 and self.__storage[i].smaller(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    i = f
                    f = self.iFather(i)
            elif self.__type == "MAX":
                while(i > 1 and self.__storage[i].bigger(self.__storage[f])):
                    self.__storage[f], self.__storage[i] = self.__storage[i], self.__storage[f]
                    i = f
                    f = self.iFather(i)
            else:
                raise ValueError('Wrong Heap type. It can be MIN or MAX')
    
    def toList(self):
        if not self.isEmpty() :
            return self.__storage[1:]
        else:
            return ["Empty Heap"]
    
    def printGraph(self):
        heap_list = self.toList()
        values = []
        for x in heap_list:
            values.append(f"{x.name}:{x.age}")
        root = build(values)
        print(root)
        
    def __str__(self) -> str:
        if not self.isEmpty() :
            heap_list = self.toList()
            str_list = ""
            for x in heap_list:
                str_list += f"{x.name}:{x.age}\n"
            return str_list
        else:
            return "Empty Heap"
```
